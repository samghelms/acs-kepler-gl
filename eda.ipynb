```python
acs_data = pd.read_csv("data/ss15husa.csv")
acs_data.head()
```

```python
acs_data[["PUMA", "ACCESS"]].head()
```

```python
acs_data_agg = acs_data.groupby(["PUMA", "ACCESS"])[["ACCESS"]].count().unstack(level=-1)
acs_data_agg
```

```python
acs_data_agg.columns = [c[0] + " = " + str(int(c[1])) for c in acs_data_agg]
```

```python
acs_data_agg
```

```python
acs_data_agg["internet coverage"] = (acs_data_agg["ACCESS = 1"] + acs_data_agg["ACCESS = 2"])\
                        / (acs_data_agg["ACCESS = 3"] + acs_data_agg["ACCESS = 1"] + acs_data_agg["ACCESS = 2"])
acs_data_agg.head()
```

```python
acs_data_dict = acs_data_agg[["internet coverage"]].to_dict(orient="index")
```

```python
len(acs_data_dict)
```

```python
# acs_data_agg
```

```python
for f in d["features"]:
    puma = int(f['properties']['PUMACE10'])
    f["properties"]["internet coverage"] = acs_data_dict[puma]["internet coverage"]
```
